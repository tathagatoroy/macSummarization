```python
attributes = ['extractiveness']
arguments = [
    ["/scratch/tathagato/naacl/zero_shot/llama_extractiveness.pkl", "llama"],
    ["/scratch/tathagato/naacl/zero_shot/mistral_extractiveness.pkl", "mistral"],
    ["/scratch/tathagato/naacl/single_attribute_sft/llama_extractiveness/results.pkl", "llama"],
    ["/scratch/tathagato/naacl/single_attribute_sft/mistral_extractiveness/results.pkl", "mistral"],
    ["/scratch/tathagato/naacl/single_attribute_dpo/llama_extractiveness/results.pkl", "llama"],
    ["/scratch/tathagato/naacl/single_attribute_dpo/mistral_extractiveness/results.pkl", "mistral"]


]
def display_extractiveness(res):
    keys = ['normal', 'high', 'fully', 'overall']
    metrics = ['fragment_density', 'coverage', 'overlap', 'f_value']
    rouge_key = ['rouge1', 'rouge2', 'rougeL']
    for met in metrics:
        for key in keys:
            print(f"{round(res['extractiveness'][key][met],2):.2f}", end = " ")
        print("")
    for key in keys:
        print(f"{round(res['extractiveness'][key]['rouge1'],2):.2f}/{round(res['extractiveness'][key]['rouge2'],2):.2f}/{round(res['extractiveness'][key]['rougeL'],2):.2f}", end = " ")

```

```python
final_res = {}
for file, model in arguments:
    print(file)
    try:
        res = evaluate(file, model=model, attributes=attributes)
        final_res[file] = {"result": res, "model": model}
    except Exception as e:
        print(f"Error evaluating {file}: {e}")
        res = None
    
```

```python
for file in final_res:
    print(file)
    model = final_res[file]["model"]
    print(model)
    display_extractiveness(final_res[file]["result"])
    print("")
    print("----------------------------------------------")
```
