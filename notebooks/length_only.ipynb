```python
import sys
sys.path.insert(0,"/home2/tathagato/summarization/MACSUM/naacl")
import os
from eval import evaluate
```

```python
attributes = ['length']
arguments = [
    ["/scratch/tathagato/naacl/zero_shot/llama_length.pkl", "llama"],
    ["/scratch/tathagato/naacl/zero_shot/mistral_length.pkl", "mistral"],
    ["/scratch/tathagato/naacl/single_attribute_sft/llama_length/results.pkl", "llama"],
    ["/scratch/tathagato/naacl/single_attribute_sft/mistral_length/results.pkl", "mistral"],
    ["/scratch/tathagato/naacl/single_attribute_dpo/llama_length/results.pkl", "llama"],
    ["/scratch/tathagato/naacl/single_attribute_dpo/mistral_length/results.pkl", "mistral"]


]
def display_length(res):
    keys = ['short' , 'normal', 'long', 'overall']
    metrics = ['summary_length', 'compression_ratio']
    rouge_key = ['rouge1', 'rouge2', 'rougeL']
    for met in metrics:
        for key in keys:
            print(f"{round(res['length'][key][met],2):.2f}", end = " ")
        print("")
    for key in keys:
        print(f"{round(res['length'][key]['rouge1'],2):.2f}/{round(res['length'][key]['rouge2'],2):.2f}/{round(res['length'][key]['rougeL'],2):.2f}", end = " ")

```

```python
final_res = {}
for file, model in arguments:
    print(file, model)
    try:
        res = evaluate(file, model=model, attributes=attributes)
        final_res[file] = {"result": res, "model": model}
    except Exception as e:
        print(f"Error evaluating {file}: {e}")
        res = None
    
```

```python
print(len(final_res.keys()))
```

```python
for file in final_res:
    print(file)
    model = final_res[file]["model"]
    print(model)
    display_length(final_res[file]["result"])
    print("")
    print("----------------------------------------------")
```
